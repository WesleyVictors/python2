#### Q4.
* Crie um dataframe `gols` com duas colunas `pa√≠s` e `total_gols` com o total de gols marcados por cada pa√≠s durante todas as copas do mundo, juntando gols em casa (`gols_1`) e gols como time visitante (`gols_2`).
* Imprima o dataframe `gols` ordenado de forma decrescente, para que os times com mais gols fiquem no topo.

Segue um exemplo ilustrativo com o formato do dataframe resultado:

```
pa√≠s        total_gols
Brazil      120
Argentina   112

```

````python
gols_casa = wc.groupby('time_1')['gols_1'].sum().reset_index()
gols_fora = wc.groupby('time_2')['gols_2'].sum().reset_index()
gols_casa.columns = ['pa√≠s', 'gols_casa']
gols_fora.columns = ['pa√≠s', 'gols_fora']


#unindo as colunas dos pa√≠ses
gols = pd.merge(gols_casa, gols_fora, on='pa√≠s', how='outer').fillna(0)
#soma dos gols casa e fora
gols['total_gols'] = gols['gols_casa'] + gols['gols_fora']

gols ['total_gols'] = gols['total_gols'].astype('int')
gols = gols[['pa√≠s', 'total_gols']].sort_values('total_gols', ascending=False)
print (gols.head(5).to_string(index=False))

````

#### Q5.
Qual pa√≠s tomou mais cart√µes amarelos somando todas as copas?

Neste exerc√≠cio voc√™ vai trabalhar com as colunas `cartao_amarelo_1` e `cartao_amarelo_2` onde cada observa√ß√£o √© uma string represetando uma lista dos cart√µes amarelos de um √∫nico jogo na forma `minuto|placar|jogador(a)`. Por exemplo:
```
['16‚Äô|1:0|Rosana G√≥mez', '20‚Äô|2:0|Gabriela Ch√°vez]
```

Recomendo criar colunas `num_cartoes_1` e `num_cartoes_2` a partir de cada coluna `cartao_amarelo_X` usando o m√©todo gen√©rico `apply` para chamar uma fun√ß√£o que voc√™ vai criar para transformar uma observa√ß√£o de cart√£o amarelo em uma contagem de elementos da lista.

Lembre de levar em considera√ß√£o que muitas observa√ß√µes s√£o `NaN`.

Em seguida fa√ßa sua m√°gica para agrupar as informa√ß√µes por pa√≠s, acumular os cart√µes de jogos em casa e jogos visitante, e produzir o resultado final como apresentado a seguir (valores inventados):

```
pa√≠s        cart√µes amarelos
Argentina   97
England     93
Australia   93
...
```

````python
import re


cartoes_casa = wc.groupby('time_1')['cartao_amarelo_1'].sum().reset_index()
cartoes_fora = wc.groupby('time_2')['cartao_amarelo_2'].sum().reset_index()
cartoes_casa.columns = ['pa√≠s', 'cartao_amarelo_1']
cartoes_fora.columns = ['pa√≠s', 'cartao_amarelo_2']

def contar_cartoes(texto):
    if pd.isnull(texto):
        return 0
    else:
        return len(re.findall(r"'(?:[^'\\]|\\.)*'", texto)) #define a separa√ß√£o dos cart√µes em['  ']


cartoes_casa['num_cartoes_1'] = cartoes_casa['cartao_amarelo_1'].apply(contar_cartoes)
cartoes_fora['num_cartoes_2'] = cartoes_fora['cartao_amarelo_2'].apply(contar_cartoes)

cartoes = pd.merge(cartoes_casa, cartoes_fora, on='pa√≠s', how='outer')
cartoes['cartoes_amarelos'] = cartoes['num_cartoes_1'] + cartoes['num_cartoes_2']
cartoes = cartoes[['pa√≠s', 'cartoes_amarelos']].sort_values('cartoes_amarelos', ascending=False).fillna(0)


cartoes['cartoes_amarelos'] = cartoes['cartoes_amarelos'].astype('int')
print(cartoes.head(5).to_string(index=False))

````

#### Q6.
Qual o top10 jogadores com mais gols em copa? Considere gols em jogo e gols de p√™nalti.

Para conseguir essa informa√ß√£o, voc√™ precisar√° trabalhar com as colunas:
```
10  gols_1_detalhes         
11  gols_2_detalhes    
14  gols_1_penalti     
15  gols_2_penalti     
```

Essas tamb√©m s√£o colunas textuais, onde cada observa√ß√µes apresenta todos os gols de uma partida separados pelo caracter `|`. Cada gol est√° na forma `'Jogador(a) ¬∑ minuto‚Äô'`. Por exemplo:
```
'Alex Morgan ¬∑ 12‚Äô|Rose Lavelle ¬∑ 20‚Äô'
```

Lembre de levar em considera√ß√£o que muitas observa√ß√µes s√£o `NaN`.

Recomendo criar um dicion√°rio √† parte, onde cada chave ser√° um jogador encontrado nessas colunas do dataframe `wc` e o valor correspondente ser√° a contagem de ocorr√™ncias desses nomes.

Em seguida basta converter o seu dicion√°rio em um dataframe e imprim√≠-lo na forma (valores n√£o s√£o inventados üòÅ):
```
jogador(a)      num_gols
Marta           17
Miroslav Klose  16
...
```

````python
# Cria um novo DataFrame vazio
gols_detalhados_df = pd.DataFrame()

# Itera sobre as colunas relevantes do DataFrame wc
for coluna in ['gols_1_detalhes', 'gols_2_detalhes', 'gols_1_penalti', 'gols_2_penalti']:
    # Separa os gols em novas linhas e adiciona ao novo DataFrame
    gols_separados = wc[coluna].str.split('|', expand=True).stack().reset_index(level=1, drop=True).rename(coluna + '_separados')
    gols_detalhados_df = pd.concat([gols_detalhados_df, wc[['data', 'time_1', 'time_2', 'copa']].join(gols_separados)], axis=0)

# Remove as linhas com valores nulos na coluna 'gols_detalhes_separados'
gols_detalhados_df = gols_detalhados_df.dropna(subset=['gols_1_detalhes_separados', 'gols_2_detalhes_separados', 'gols_1_penalti_separados', 'gols_2_penalti_separados'], how='all')

def remover_apos_ponto(texto):
  if pd.isna(texto):
    return texto
  else:
    return texto.split('¬∑')[0].strip()

# Aplica a fun√ß√£o √†s colunas criadas
for coluna in ['gols_1_detalhes_separados', 'gols_2_detalhes_separados', 'gols_1_penalti_separados', 'gols_2_penalti_separados']:
    gols_detalhados_df[coluna] = gols_detalhados_df[coluna].apply(remover_apos_ponto)

display(gols_detalhados_df)
````

````python
# Cria o DataFrame 'artilharia' com as colunas desejadas
artilharia = gols_detalhados_df[['gols_1_detalhes_separados', 'gols_2_detalhes_separados', 'gols_1_penalti_separados', 'gols_2_penalti_separados']].copy()

# Junta as quatro colunas em uma √∫nica coluna, removendo valores NaN
artilharia['todos_gols'] = artilharia.stack().reset_index(drop=True) #stack junta as colunas, drop true remove os indices
artilharia = artilharia[['todos_gols']].dropna()  # Seleciona apenas a coluna 'todos_gols' e remove linhas com NaN

#display(artilharia)
````

````python
# Conta as ocorr√™ncias de cada nome e cria um novo DataFrame
contagem_gols = artilharia['todos_gols'].value_counts().reset_index()

# Renomeia as colunas para 'jogador(a)' e 'num_gols'
contagem_gols.columns = ['jogador(a)', 'num_gols']

# Exibe o DataFrame com a contagem de gols
display(contagem_gols)
````
